```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.random.uniform(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.1, 100)
```

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Training set')
plt.show()
```

```python
x_test = np.array([7, 8, 9, 10, 11, 12, 13, 14, 15])
y_test = np.sin(x_test)
```

```python
plt.scatter(x, y, label='training set')
plt.scatter(x_test, y_test, label='actual values (test set)', marker='o', s=100)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Training and test sets')
plt.legend(loc='lower left')
plt.show()
```

```python
from sklearn.linear_model import Ridge
from sklearn.kernel_ridge import KernelRidge
from sklearn.gaussian_process.kernels import ExpSineSquared
```

```python
linear_model = Ridge()
linear_model.fit(x.reshape(len(x), 1), y)

kernel_rbf = KernelRidge(kernel='rbf')
kernel_rbf.fit(x.reshape(len(x), 1), y)

kernel_expsine = KernelRidge(kernel=ExpSineSquared(periodicity=2*np.pi))
kernel_expsine.fit(x.reshape(len(x), 1), y)

plt.scatter(x, y)
plt.scatter(x_test, y_test,
            label='Actual values', marker='o', s=100, edgecolors='black')

plt.scatter(x_test, linear_model.predict(x_test.reshape(len(x_test), 1)),
            label='Linear model', marker='o', s=100, edgecolors='black')

plt.scatter(x_test, kernel_rbf.predict(x_test.reshape(len(x_test), 1)),
            label='RBF Kernel', marker='o', s=100, edgecolors='black')

plt.scatter(x_test, kernel_expsine.predict(x_test.reshape(len(x_test), 1)),
            label='ExpSine Kernel', marker='o', s=100, edgecolors='black')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Training and test sets')
plt.legend(loc='lower left')
plt.show()
```
